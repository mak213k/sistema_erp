```python
import pandas as pd
import ipywidgets as widgets
from IPython.display import display, clear_output

# --- CONFIGURA√á√ïES INICIAIS ---
setores = ["Montagem", "Almoxarifado", "Pintura", "Serralheria"]
estoque_dados = {} # Dicion√°rio para armazenar os objetos das pe√ßas

class Peca:
    def __init__(self, codigo, descricao, quantidade):
        self.codigo = codigo
        self.descricao = descricao
        self.quantidade = quantidade
        self.status = "Em estoque"
        self.setor_atual = "Almoxarifado"
        self.enviado = False

    def resumo(self):
        return {
            "C√≥digo": self.codigo,
            "Descri√ß√£o": self.descricao,
            "Quantidade": self.quantidade,
            "Status": self.status,
            "Setor Atual": self.setor_atual
        }

# --- FUN√á√ïES DE INTERA√á√ÉO ---
def atualizar_tabela():
    with out_tabela:
        clear_output(wait=True)
        if estoque_dados:
            df = pd.DataFrame([p.resumo() for p in estoque_dados.values()])
            display(df)
        else:
            print("Nenhuma pe√ßa cadastrada.")

def cadastrar_peca(b):
    cod = txt_cod.value.strip()
    desc = txt_desc.value.strip()
    qtd = num_qtd.value
    
    if cod and desc:
        estoque_dados[cod] = Peca(cod, desc, qtd)
        # Atualiza o dropdown de sele√ß√£o de pe√ßas
        sel_peca_mov.options = list(estoque_dados.keys())
        atualizar_tabela()
        # Limpa os campos
        txt_cod.value = ""
        txt_desc.value = ""
        num_qtd.value = 1
    else:
        with out_tabela:
            print("‚ö†Ô∏è Preencha c√≥digo e descri√ß√£o!")

def realizar_movimentacao(b):
    cod = sel_peca_mov.value
    acao = sel_acao.value
    setor = sel_setor.value
    
    if cod in estoque_dados:
        peca = estoque_dados[cod]
        
        if acao == "Enviar":
            peca.setor_atual = setor
            peca.enviado = True
            peca.status = f"üöö Em tr√¢nsito para {setor}"
        
        elif acao == "Receber":
            if peca.enviado:
                peca.enviado = False
                peca.status = f"‚úÖ Dispon√≠vel em {peca.setor_atual}"
            else:
                with out_tabela: print(f"‚ö†Ô∏è Pe√ßa {cod} n√£o foi enviada para ser recebida.")
        
        elif acao == "Dar Baixa":
            peca.status = "üö´ Baixada/Finalizada"
            peca.setor_atual = "---"
            
        atualizar_tabela()

# --- CRIA√á√ÉO DOS WIDGETS (INTERFACE) ---

# Widgets de Cadastro
txt_cod = widgets.Text(description="C√≥digo:", placeholder="P001")
txt_desc = widgets.Text(description="Descri√ß√£o:", placeholder="Ex: Chapa de A√ßo")
num_qtd = widgets.IntText(description="Qtd:", value=1)
btn_add = widgets.Button(description="Cadastrar Pe√ßa", button_style='success', icon='plus')
btn_add.on_click(cadastrar_peca)

# Widgets de Movimenta√ß√£o
sel_peca_mov = widgets.Dropdown(description="Pe√ßa:", options=[])
sel_acao = widgets.Dropdown(description="A√ß√£o:", options=["Enviar", "Receber", "Dar Baixa"])
sel_setor = widgets.Dropdown(description="Para Setor:", options=setores)
btn_exec = widgets.Button(description="Executar", button_style='primary', icon='refresh')
btn_exec.on_click(realizar_movimentacao)

# Sa√≠da da Tabela
out_tabela = widgets.Output()

# --- LAYOUT FINAL ---
layout_cadastro = widgets.VBox([
    widgets.HTML("<h2>üì¶ Cadastro de Pe√ßas</h2>"),
    widgets.HBox([txt_cod, txt_desc, num_qtd, btn_add])
])

layout_movimento = widgets.VBox([
    widgets.HTML("<h2>‚öôÔ∏è Movimenta√ß√£o e Status</h2>"),
    widgets.HBox([sel_peca_mov, sel_acao, sel_setor, btn_exec])
])

display(layout_cadastro)
display(widgets.HTML("<br>"))
display(layout_movimento)
display(widgets.HTML("<hr>"))
display(out_tabela)

# Inicializar com a tabela vazia
atualizar_tabela()
```
